```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 50)
```

```python
df_diabetes = pd.read_csv('/content/Diabetes_Final.csv')
df_diarrhea = pd.read_csv(r'/content/Diarrheal diseases.csv', encoding='latin-1')
df_Respiratory_Upper_lower_infection = pd.read_csv('/content/Respiratory_Upper_lower_infection.csv', encoding='latin-1')
df_covid = pd.read_csv('/content/covid.csv')
df_dengue = pd.read_csv('/content/dengue.csv')
df_heart_diseases = pd.read_csv('/content/heart_diseases.csv',  encoding='latin-1')
```

```python
df_diabetes['condition'] = 'diabetes'
df_diarrhea['condition'] = 'diarrhea'
df_Respiratory_Upper_lower_infection['condition'] = 'respiratory infection'
df_covid['condition'] = 'covid'
df_dengue['condition'] = 'dengue'
df_heart_diseases['condition'] = 'heart diseases'

df_diabetes.head()
```

```python
merged_df = pd.concat([df_diabetes, df_diarrhea, df_Respiratory_Upper_lower_infection, df_covid, df_dengue, df_heart_diseases], ignore_index=True, sort=False)
merged_df.head()
```

```python
merged_df.to_csv('merged_data.csv', index = False)
```
